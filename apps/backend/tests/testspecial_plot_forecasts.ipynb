```python
# Ala Archa
dataf = data[data['code']==15194]
# Toktogul inflow
#dataf = data[data['code']==16936]
print(dataf)
```

```python
# Read in data from excel sheet
# print current working directory
print("Current Working Directory " , os.getcwd())
# Get a list of strings from 2000 to 2022
years = [str(i) for i in range(2000, 2023)]
# Read in data from excel where the sheet names are stored in the years list and concatenate all the dataframes
# Ala Archa
reference = pd.concat([pd.read_excel("../../../../pentad_forecast_data/daily_runoff/15194_Ала-Арча 2000-2023.xlsx", sheet_name=year) for year in years])
# Toktogul inflow
#reference = pd.concat([pd.read_excel("../../../../pentad_forecast_data/daily_runoff/16936_приток в токтогул 2000-2023г.xlsx", sheet_name=year) for year in years])
# Rename the columns to date and runoff_m3s
reference.columns = ['date', 'runoff_m3s']
reference['date'] = pd.to_datetime(reference['date'], format='%Y-%m-%d')
print(reference)
```

```python
# Calculate the predictors based on the reference data
forecast_days = [5, 10, 15, 20, 25]  # plus the last day of the month
# in the reference dataframe, identify if the date is in the forecast_days list
# or if it is the last day of the month
# If it is, then create a new column in the dataframe with the value of 1,
# otherwise 0.
reference['is_forecast_date'] = reference['date'].apply(
    lambda x: 1 if x.day in forecast_days or x.day == x.days_in_month else 0)
# Create a new dataframe by summing the runoff_m3s values from dates where
# is_forecast_date is 1 minus 1 day to dates where is_forecast_date is 1 minus
# 3 days.
# This is the predictor for the 5 day forecast.
reference['predictor_5'] = reference['runoff_m3s'].shift(1).rolling(3).sum()
# Calculate the 5 day average of runoff_m3s values from dates where
# is_forecast_date is 1 plus 1 day to dates where is_forecast_date is 1 plus 5 days.
# This is the target for the 5 day forecast.
reference['target_5'] = reference['runoff_m3s'].shift(-5).rolling(5).mean()
print(reference.head(180).tail(25))
```

```python
# Now merge the predictor_5 column from the reference dataframe to the dataf
# dataframe by matching the date columns.
dataf = dataf.merge(reference[['date', 'predictor_5', 'target_5']], on='date', how='left')
# Add a column with the difference between predictor and predictor_5
dataf['diff'] = dataf['predictor'] - dataf['predictor_5']
# Set all diff values that are smaller than e-6 to 0
dataf.loc[dataf['diff'] < 1e-6, 'diff'] = 0
print(dataf)
```

```python
p = hv.Curve(dataf, kdims='date', vdims=['diff']).opts(tools=['hover']) *\
hv.Curve(dataf, kdims='date', vdims=['predictor']).opts(tools=['hover']) *\
hv.Curve(dataf, kdims='date', vdims=['predictor_5']).opts(tools=['hover'])
#hv.Curve(dataf, kdims='date', vdims=['fc_qmin']) *\
#hv.Curve(dataf, kdims='date', vdims=['fc_qexp']) *\
#hv.Curve(dataf, kdims='date', vdims=['fc_qmax'])
p.opts(width=600)
p
```

```python
# calculate forecast accuracy based on dataf

# calculate the v_factor based on fq_exp and fq_max
dataf['v_factor'] = dataf['fc_qmax'] - dataf['fc_qexp']
# Test if fc_qexp was within acceptable range of the target plus minus v_factor
# If it is, set accuracy to 1, otherwise 0
dataf['accuracy'] = dataf.apply(lambda x: 1 if x['target_5'] - x['v_factor'] <= x['fc_qexp'] <= x['target_5'] + x['v_factor'] else 0, axis=1)
print(dataf[['fc_qexp', 'v_factor', 'target_5', 'accuracy']])

```

```python
# Now we add a column for the pentad of the year. The pentad of the year is an
# integer between 1 and 73 that represents the pentad of the year. The first
# pentad of the year is from 1st January to 5th January. The second pentad of
# the year is from 6th January to 10th January and so on. The last pentad of the
# year is from 26th December to 31st December.
dataf['pentad'] = (dataf['date']+dt.timedelta(days=1.0)).dt.dayofyear // 5 + 1
print(dataf[['date', 'pentad']].head(380).tail(40))
```

```python
# Now we only look at the data from 1st pentad in 2010 to the last pentad in 2022.
# We do this by filtering the data based on the date column.
datafs = dataf[(dataf['date'].dt.year >= 2010) & (dataf['date'].dt.year <= 2022)]
print(datafs[['date', 'pentad']].head(2))
print(datafs[['date', 'pentad']].tail(2))
```

```python
# By pentad, calculate the average accuracy.
# This is done by grouping the data by the pentad column and calculating the
# mean of the accuracy column.
accuracy = datafs.groupby('pentad')['accuracy'].mean()*100.0

print(accuracy)
```

```python
# Plot an interactive line plot of the accuracy by pentad
p = hv.Curve(accuracy, kdims='pentad', vdims=['accuracy']).\
    opts(tools=['hover'], interpolation='steps-mid')
p.opts(width=800, ylim=(0, 110), show_grid=True)
p
```
