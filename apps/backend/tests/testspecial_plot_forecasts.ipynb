```python
import datetime as dt
import pandas as pd
import holoviews as hv
import os

hv.extension('bokeh')

# Read in csv file
data = pd.read_csv("../../internal_data/forecasts_pentad_kghm_full_hindcast.csv")
#data = pd.read_csv("~/hydrosolutions Dropbox/Bea martibeatrice@gmail.com/SAPPHIRE_Central_Asia_Technical_Work/code/pentadal_forecasting_system/hindcast_results/20240326/internal_data/forecasts_pentad_kghm_full_hindcast.csv")
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
print(data)
```

```python
dataf = data[data['code']==15194]
print(dataf)
```

```python
# Read in data from excel sheet
# print current working directory
print("Current Working Directory " , os.getcwd())
# Get a list of strings from 2000 to 2022
years = [str(i) for i in range(2000, 2023)]
# Read in data from excel where the sheet names are stored in the years list and concatenate all the dataframes
reference = pd.concat([pd.read_excel("../../../../pentad_forecast_data/daily_runoff/15194_Ала-Арча 2000-2023.xlsx", sheet_name=year) for year in years])
# Rename the columns to date and runoff_m3s
reference.columns = ['date', 'runoff_m3s']
reference['date'] = pd.to_datetime(reference['date'], format='%Y-%m-%d')
print(reference)
```

```python
# Calculate the predictors based on the reference data
forecast_days = [5, 10, 15, 20, 25]  # plus the last day of the month
# in the reference dataframe, identify if the date is in the forecast_days list
# or if it is the last day of the month
# If it is, then create a new column in the dataframe with the value of 1,
# otherwise 0.
reference['is_forecast_date'] = reference['date'].apply(
    lambda x: 1 if x.day in forecast_days or x.day == x.days_in_month else 0)
# Create a new dataframe by summing the runoff_m3s values from dates where
# is_forecast_date is 1 minus 1 day to dates where is_forecast_date is 1 minus
# 3 days.
# This is the predictor for the 5 day forecast.
reference['predictor_5'] = reference['runoff_m3s'].shift(1).rolling(3).sum()
print(reference.head(180).tail(25))
```

```python
# Now merge the predictor_5 column from the reference dataframe to the dataf
# dataframe by matching the date columns.
dataf = dataf.merge(reference[['date', 'predictor_5']], on='date', how='left')
# Add a column with the difference between predictor and predictor_5
dataf['diff'] = dataf['predictor'] - dataf['predictor_5']
# Set all diff values that are smaller than e-6 to 0
dataf.loc[dataf['diff'] < 1e-6, 'diff'] = 0
print(dataf)
```

```python
p = hv.Curve(dataf, kdims='date', vdims=['diff']).opts(tools=['hover']) *\
hv.Curve(dataf, kdims='date', vdims=['predictor']).opts(tools=['hover']) *\
hv.Curve(dataf, kdims='date', vdims=['predictor_5']).opts(tools=['hover'])
#hv.Curve(dataf, kdims='date', vdims=['fc_qmin']) *\
#hv.Curve(dataf, kdims='date', vdims=['fc_qexp']) *\
#hv.Curve(dataf, kdims='date', vdims=['fc_qmax'])
p.opts(width=600)
p
```
